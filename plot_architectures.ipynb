```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the style using seaborn
sns.set_style("whitegrid", {"grid.color": ".9"})
sns.set_context("paper", font_scale=1, rc={"axes.linewidth": 1.5})

# Data for plotting
x = [24, 25, 27, 42, 54]
y = [12.39, 12.57, 12.96, 13.8, 13.9]
labels = ["InceptionV3", "EfficientNetB0", "EfficientNetB1", "DenseNet169", "Xception"]

# Plotting the Lorenz curve
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
plt.scatter(x, y, color='#175D8D', s=100)  # Increase marker size for visibility
plt.plot(x, y, color='#1E76B4', linewidth=2.5)

# Labeling the points
for i, txt in enumerate(labels):
    to_add = [4.5, 5.25, 5.25, 5.15, 3.55]
    ax.text(x[i]+to_add[i], y[i], txt, fontsize=12, ha='right', va='bottom',
            bbox=dict(boxstyle="round,pad=0.3", edgecolor='grey', facecolor='white', alpha=1))

# Enhancing plot aesthetics
plt.xlabel('Number of network parameters (millions)', fontsize=18)
plt.ylabel('RMSE over all categories [knots]', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Adding subtle artistic elements
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
```
